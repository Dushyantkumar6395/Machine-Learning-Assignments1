```python
# Q1. What is the Probability density function?
# The Probability Density Function (PDF) is a concept used in probability theory and statistics to describe the
# probability distribution of a continuous random variable. Unlike the Probability Mass Function (PMF), which is 
# used for discrete random variables, the PDF applies to continuous random variables.
```

```python
# Q2. What are the types of Probability distribution?
# There are several types of probability distributions, each used to model different types of random variables:

# 1. **Discrete Distributions**: These distributions model discrete random variables where outcomes can only take distinct values, such as integers or counts. Examples include the Binomial distribution, Poisson distribution, and Geometric distribution.

# 2. **Continuous Distributions**: These distributions model continuous random variables where outcomes can take any value within a range, often described by a Probability Density Function (PDF). Examples include the Normal (Gaussian) distribution, Exponential distribution, and Uniform distribution.

# 3. **Mixed Distributions**: These distributions combine elements of both discrete and continuous distributions, where some outcomes are discrete and others are continuous. Examples include the Compound Poisson distribution and Log-Normal distribution.

# 4. **Multivariate Distributions**: These distributions describe the joint probability distribution of multiple random variables simultaneously. Examples include the Multinomial distribution and Multivariate Normal distribution.

# 5. **Special Distributions**: These are distributions with specific characteristics or applications, often arising in specialized areas of statistics and mathematics. Examples include the Chi-Square distribution, Student's t-distribution, and Beta distribution.

# Understanding the characteristics and properties of each type of probability distribution is crucial for accurately modeling and analyzing real-world data and phenomena in various fields such as finance, engineering, and the natural sciences.
```

```python
# Q3. Write a Python function to calculate the probability density function of a normal distribution with
# given mean and standard deviation at a given point.

import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution at a given point x.

    Parameters:
    - x: The point at which to evaluate the PDF.
    - mean: The mean (or expectation) of the normal distribution.
    - std_dev: The standard deviation (or square root of the variance) of the normal distribution.

    Returns:
    - pdf_value: The PDF value at point x.
    """
    pdf_value = (1 / (std_dev * math.sqrt(2 * math.pi))) * math.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))
    return pdf_value

# Example usage:
mean = 0
std_dev = 1
x = 0.5
pdf_value = normal_pdf(x, mean, std_dev)
print(f"PDF at x = {x}: {pdf_value:.4f}")
```

```python
# # Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
# # distribution can be applied.

# The Binomial distribution is characterized by the following properties:

# 1. **Discrete Distribution**: It models the number of successes \( k \) in a fixed number \( n \) of independent Bernoulli trials, where each trial has a constant probability \( p \) of success.
   
# 2. **Parameters**: It is defined by two parameters: \( n \) (number of trials) and \( p \) (probability of success in each trial).

# 3. **Probability Mass Function (PMF)**: The PMF of the Binomial distribution is given by \( P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} \), where \( X \) is the random variable representing the number of successes.

# 4. **Mean and Variance**: The mean \( E(X) = np \) and variance \( Var(X) = np(1-p) \).

# 5. **Examples**: 
#    - **Example 1**: The number of heads obtained when flipping a coin \( n \) times, where \( p \) is the probability of getting heads.
#    - **Example 2**: The number of defective items in a batch of \( n \) items produced, where \( p \) is the probability of a defect occurring.

# The Binomial distribution is widely applicable in scenarios involving repeated independent trials with two possible outcomes (success or failure), making it suitable for modeling various real-world phenomena such as coin flips, manufacturing quality control, and medical testing outcomes.
```

```python
# Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
# and plot a histogram of the results using matplotlib.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10  # Number of trials
p = 0.4  # Probability of success

# Generate random sample
sample_size = 1000
sample = np.random.binomial(n, p, sample_size)

# Plot histogram
plt.figure(figsize=(8, 6))
plt.hist(sample, bins=np.arange(0, n+2)-0.5, edgecolor='black', alpha=0.7)
plt.xticks(range(n+1))
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title(f'Binomial Distribution: n={n}, p={p}, Sample Size={sample_size}')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
# with given mean at a given point.

import math

def poisson_cdf(k, mean):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution at a given point k.

    Parameters:
    - k: The point at which to evaluate the CDF.
    - mean: The mean (or lambda) of the Poisson distribution.

    Returns:
    - cdf_value: The CDF value at point k.
    """
    cdf_value = 0.0
    for i in range(k + 1):
        cdf_value += math.exp(-mean) * (mean ** i) / math.factorial(i)
    return cdf_value

# Example usage:
mean = 2.5
k = 3
cdf_value = poisson_cdf(k, mean)
print(f"CDF at k = {k}: {cdf_value:.4f}")
```

```python
# Q7. How Binomial distribution different from Poisson distribution?
# The Binomial and Poisson distributions differ in several key aspects:

# 1. **Nature of Events**: The Binomial distribution models the number of successes in a fixed number of independent trials with a constant probability of success, denoted by \( n \) (number of trials) and \( p \) (probability of success per trial). In contrast, the Poisson distribution models the number of events occurring in a fixed interval of time or space, where events happen independently at a constant rate \( \lambda \).

# 2. **Parameters**: The Binomial distribution is characterized by two parameters, \( n \) and \( p \), while the Poisson distribution is characterized by a single parameter \( \lambda \).

# 3. **Distribution Type**: The Binomial distribution is discrete, as it deals with counts of discrete events (e.g., number of heads in coin flips), while the Poisson distribution is also discrete but typically deals with counts of rare events over a continuous domain (e.g., number of phone calls received in a day).

# 4. **Limiting Behavior**: As \( n \), the number of trials in the Binomial distribution, becomes large and \( p \), the probability of success, becomes small such that \( \lambda = np \) remains constant, the Binomial distribution approaches the Poisson distribution. This is known as the Poisson limit of the Binomial distribution.

# 5. **Application**: The Binomial distribution is commonly used in scenarios involving a fixed number of trials with known probabilities, such as quality control in manufacturing or success/failure experiments. The Poisson distribution is used when dealing with events that occur randomly over time or space, such as arrival of customers at a store or occurrences of accidents in a city.

# Understanding these distinctions helps in choosing the appropriate distribution for modeling real-world phenomena based on the nature of the events and the available parameters.
```

```python
# Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
# sample mean and variance.

import numpy as np

# Parameters
mean = 5
sample_size = 1000

# Generate random sample from Poisson distribution
sample = np.random.poisson(mean, sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample, ddof=0)  # ddof=0 for population variance

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")
```

```python
# Q10. In normal distribution with respect to mean position, where does the least frequent data appear?
# In a normal distribution, also known as a Gaussian distribution:
  
# 1. **Central Tendency**: The mean (\(\mu\)) represents the center of the distribution, where the data are most concentrated.
  
# 2. **Least Frequent Data**: The least frequent data points appear in the tails of the distribution, farther away from the mean.
  
# 3. **Tail Ends**: Specifically, the data points that are several standard deviations (\(\sigma\)) away from the mean (\(\mu\)) are the least frequent.
  
# 4. **Probability Density**: According to the 68-95-99.7 rule (or the empirical rule), about 68% of data points lie within one standard deviation of the mean, leaving approximately 32% of the data in the tails.
  
# 5. **Extremes**: Therefore, the least frequent data points appear at the extremes of the distribution, in the tails beyond two or three standard deviations from the mean, where the probability density decreases significantly.
```
